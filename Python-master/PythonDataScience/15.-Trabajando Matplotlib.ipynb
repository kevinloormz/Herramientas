```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/dell/OneDrive - Universidad de Guayaquil/Analista de datos - Linkedin Learning/base_datos_2008.csv",nrows=1000000)
```

```python
#a√±adimos una semilla 
np.random.seed()
```

```python
df=df.sample(frac=1).head(100)
```

```python
#creamos el grafico 
plt.scatter(x=df.DayofMonth,y=df.ArrDelay,s=df.Distance)
```

```python
#modifiamos el grafico con nuevos parametros ejepmlo alpha que hace que las burbujas sean opacas
#c=6>sabado, 7>domingo
plt.scatter(x=df.DayofMonth,y=df.ArrDelay,s=df.Distance,alpha=0.3,c=df.DayOfWeek.isin([6,7]))
plt.title("Retrasos vuelos EEUU")
plt.xlabel("Dia del mes")
plt.ylabel("Retraso al llegar")
plt.ylim([0,150])
plt.xticks([0,15,30]) #podemos agregar solo tres o mas
plt.text(x=28,y=120,s="Mi vuelo")
plt.show()
```
