```python
import pandas as pd 
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
df = pd.read_csv('base_datos_2008.csv', nrows=1e4)
```

```python
newdf=df[["AirTime","DepDelay"]].dropna()
```

```python
clstr=AgglomerativeClustering(n_clusters=5)
clstr.fit(newdf)
```

```python
#podemos visualizar que tenemos dos grupos bien cerca 
plt.scatter(newdf.AirTime,newdf.DepDelay,c=clstr.fit_predict(newdf))
plt.show()
```

```python
#vamos a cargar 1 millon de registros ejemplo pueden ser nuevos clientes y vamos a predecir a que grupo pertenecen 
df = pd.read_csv('base_datos_2008.csv', nrows=2e4)[["AirTime","DepDelay"]].dropna()
```
