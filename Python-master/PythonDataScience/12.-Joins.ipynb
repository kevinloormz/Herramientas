```python
import pandas as pd
```

```python
consumidores=[("A","Movil"),("B","Movil"),("A","Portatil"),("A","Tablet"),
              ("B","Tablet"),("C","Portatil"),("D","Smart"),("E","Consoola")]
con_labels=["consumidor","producto"]
```

```python
con_df=pd.DataFrame(consumidores,columns=con_labels)
con_df
```

```python
productores=[("a","Movil"),("a","Smart"),("a","Tablet"),("b","Portatil"),
              ("c","Sobremesa"),("c","Portatil")]
pro_labels=["productor","producto"]
```

```python
pro_df=pd.DataFrame(productores,columns=pro_labels)
pro_df
```

```python
pd.merge(con_df,pro_df,on="producto",how="outer")
```

```python
pd.merge(con_df,pro_df,on="producto",how="inner")
```

```python
pd.merge(con_df,pro_df,on="producto",how="right")
```

```python
pd.merge(con_df,pro_df,on="producto",how="left")
```
