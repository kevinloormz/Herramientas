```python
from sklearn import preprocessing
import pandas as pd 
import numpy as np
df = pd.read_csv("C:/Users/dell/OneDrive - Universidad de Guayaquil/Analista de datos - Linkedin Learning/base_datos_2008.csv",nrows=100000)
```

```python
df.columns
```

```python
#eliminamos los datos faltantes de las columnas que nos interesan 
df=df[["ArrDelay","DepDelay","Distance","AirTime"]].dropna()
```

```python
df
```

```python
#algunos de los metodos de ml necesitan que los datos esten escalados 
#los datos escalados se centran en el cero 
x_scaled=preprocessing.scale(df)
x_scaled
```

```python
#que todas las variables esten en una misma escala es muy util por varios motivos 
x_scaled.std(axis=0)
```
