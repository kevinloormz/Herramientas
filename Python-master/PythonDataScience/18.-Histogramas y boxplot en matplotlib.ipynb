```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/dell/OneDrive - Universidad de Guayaquil/Analista de datos - Linkedin Learning/base_datos_2008.csv")
```

```python
df.columns
```

```python
df.dropna(inplace=True,subset=["ArrDelay","DepDelay","Distance"])
```

```python
sns.distplot(df.Distance)
```

```python
#eliminamos la densidad con kde
sns.distplot(df.Distance,kde=False)
```

```python
#podemos añadir el numero de bins
sns.distplot(df.Distance,kde=False,bins=100)
```

```python
sns.kdeplot(df.ArrDelay)
sns.kdeplot(df.DepDelay)
plt.xlim([-300,300])
```

```python
#vamos a hacer un subset 

df2=df[df.Origin.isin(["ATL","HOU","IND"])].sample(frac=1).head(5000)
```

```python
df2.head()
```

```python
#los rombos son los outliers 
#estos graficos sirven para ver donde estan concentrados nuestros datos y sacar conclusiones de exploratoria
sns.boxplot(x=df2.DepDelay,y=df2.Origin,data=df2)
plt.xlim([-20,150])
```
