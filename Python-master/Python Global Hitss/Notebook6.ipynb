One Hot Encoding

```
import pandas as pd
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

```

```
url="https://www.w3schools.com/python/cars.csv"
df=pd.read_csv(url)
df.head()
```

```
Car_dummies=pd.get_dummies(df.Car)
Car_dummies
```

```
Car_dummies=pd.get_dummies(df.Car)
Model_dummies=pd.get_dummies(df.Model)
X=pd.concat([df[["Volume",	"Weight"]], Car_dummies, Model_dummies], axis=1)
print(X.head())
y=df['CO2']
print(y.head())
print(X.describe())
print(y.describe())
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```

regr=linear_model.LinearRegression()
regr.fit(X_train,y_train)
predict_y_test_regr=regr.predict(X_test)
print(predict_y_test_regr)
print(mean_squared_error(predict_y_test_regr,y_test))
print(r2_score(predict_y_test_regr,y_test))
```

RandomForest

```
#RandomForest
modelRF= RandomForestClassifier(n_estimators=10000)
modelRF.fit(X_train, y_train)
predict_y_test_RF=modelRF.predict(X_test)
print(predict_y_test_RF)
print(mean_squared_error(predict_y_test_RF,y_test))
print(r2_score(predict_y_test_RF,y_test))
```

```
#kNN
modelkNN=KNeighborsClassifier(n_neighbors=6)
modelkNN.fit(X_train, y_train)
predict_y_test_kNN=modelkNN.predict(X_test)
print(predict_y_test)
print(mean_squared_error(predict_y_test_kNN,y_test))
print(r2_score(predict_y_test_kNN,y_test))
```

```
#Naive Bayes
modelNB=GaussianNB()
modelNB.fit(X_train, y_train)
predict_y_test_NB=modelNB.predict(X_test)
print(predict_y_test_NB)
print(mean_squared_error(predict_y_test_NB,y_test))
print(r2_score(predict_y_test_NB,y_test))
```

```
#SVM
modelsvm = svm.SVC()
modelsvm.fit(X_train, y_train)
predict_y_test_svm=modelsvm.predict(X_test)
print(predict_y_test_svm)
print(mean_squared_error(predict_y_test_svm,y_test))
print(r2_score(predict_y_test_svm,y_test))
```
